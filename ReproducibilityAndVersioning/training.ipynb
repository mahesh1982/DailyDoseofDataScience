```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# For reproducibility\n
torch.manual_seed(0)
```

```python
# 1. Define a simple model\n",
class Net(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(Net, self).__init__()
        self.linear1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_dim, 1)  # output 1 logit for binary classification

    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        return x
```

```python
model = Net(input_dim=20, hidden_dim=10)
```

```python
criterion = nn.BCEWithLogitsLoss()  # for binary classification\n",
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
X = torch.randn(100, 20)
y = torch.randint(0, 2, (100, 1)).float()
```

```python
model.train()
num_epochs = 5
for epoch in range(1, num_epochs+1):
    # Forward pass\n",
    outputs = model(X)
    loss = criterion(outputs, y)
    # Backward and optimize\n",
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    # Logging\n",
    print(f"Epoch [{epoch}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
torch.save(model.state_dict(), "model_weights.pth")
print("Model weights saved.")
```

```python
model2 = Net(input_dim=20, hidden_dim=10)
model2.load_state_dict(torch.load("model_weights.pth"))
model2.eval()
```
