```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
model.encode("This is Maheswara Reddy")
```

```python
s1 = "I hate this product"
s2 = "I love this product"
```

```python
e1 = model.encode(s1).reshape(1,-1)
e2 = model.encode(s2).reshape(1,-1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(e1, e2)
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch
```

```python
name = 'distilbert-base-uncased'
tokenizer = DistilBertTokenizer.from_pretrained(name)
model = DistilBertModel.from_pretrained(name)
```

```python
i_s1 = tokenizer(s1, return_tensors='pt')
i_s2 = tokenizer(s2, return_tensors='pt')
```

```python
with torch.no_grad():
    o1 = model(**i_s1)
    o2 = model(**i_s2)
```

```python
e1 = o1.last_hidden_state.mean(dim=1).numpy()
e2 = o2.last_hidden_state.mean(dim=1).numpy()
```

```python
cosine_similarity(e1, e2)
```
